# 8-3. 결측값 처리하기
* 평균값으로 넣기
* 최빈값으로 넣기
* 특정 값으로 넣기
* 삭제하기

```python
# 1. 평균값으로 채우기
height_mean = round(d['키'].mean(),1)
height_mean
```

```python
#수동 대입 (원본에 영향 x)
d['키'] = d['키'].fillna(height_mean)
d
```

```python
#바로 대입(원본에 영향 O)
d['키'].fillna(height_mean,inplace=True)
d
```

```python
# 2. 최빈값으로 채우기
d = pd.read_csv('korean-idol.csv')
d
```

```python
height_mode = d['키'].mode()
height_mode
```

```python
d['키'].fillna(height_mode[0],inplace=True)
d
```

```python
# 3. 특정값으로 채우기
d=pd.read_csv('korean-idol.csv')
d['키'].fillna(-1,inplace=True)
d
```

```python
d=pd.read_csv('korean-idol.csv')
d
```

```python
# 4. 결측값이 있는 행 삭제하기
d.dropna(inplace=True) # axis=0,how=any
d
```

```python
d.isnull().sum()
```

```python
d=pd.read_csv('korean-idol.csv')
d
```

```python
#결측값이 있는 열 삭제
d.dropna(axis=1,inplace=True)
d
```
