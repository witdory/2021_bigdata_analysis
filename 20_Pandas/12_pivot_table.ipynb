# 12. pivot_table, groupby 사용하기

```python
import numpy as np
import pandas as pd
```

```python
d = pd.read_csv('korean-idol.csv')
d.head()
```

## 12-1. pivot_table
* 피벗테이블은 엑셀 피벗테이블과 동일함
* 데이터 열들 중에서 행 인덱스로 보여줄 열과 열 인덱스로 보여줄 열을 선택하여 데이터를 그룹해서 보여줌
* 예) 연도별로 남녀 출생자수 합계

```python
# 그룹과 성별에 따른 키 평균 구하기
pd.pivot_table(d,index='그룹',columns='성별',values='키',aggfunc=np.mean)
```

```python
d.loc[d['그룹']=='방탄소년단','키'].mean()
```

```python
# 그룹과 혈액형에 따른 브랜드 평판지수의 평균 구하기
pd.pivot_table(d,index='그룹',columns='혈액형',values='브랜드평판지수')
```

```python
d.loc[(d['그룹'] == '방탄소년단')&(d['혈액형']=='A'),'브랜드평판지수'].mean()
```

```python
# Multi-index
#그룹내 혈액형과 성별에 따른 브랜드 평판지수의 평균 구하기
pd.pivot_table(d,index=['그룹','혈액형'], columns='성별', values='브랜드평판지수')
```

## 12-2.groupby
* groupby(['그룹핑열']).aggfunc()
* groupby(['그룹핑열'])['추출열'].aggfunc()
* pivot_table과 groupby는 보여주는 형식만 차이가 남
* 예)일일 주식 거래량 합계

```python
d.groupby('그룹').mean()
```

```python
d.groupby('그룹')['키'].mean()
```

```python
pd.DataFrame(d.groupby('그룹')['키']).mean()
```

```python
# 그룹과 성별에 따른 키 평균 구하기
pd.DataFrame(d.groupby(['그룹','성별'])['키'].mean())
```

```python
# 그룹과 혈액형에 따른 브랜드 평판지수의 평균 구하기
pd.DataFrame(d.groupby(['그룹','혈액형'])['브랜드평판지수'].mean())
```

```python
# 그룹 내 혈액형과 성별에 따른 브랜드 평판 지수의 평균 구하기
pd.DataFrame(d.groupby(['그룹','혈액형','성별'])['브랜드평판지수'].mean())
```
