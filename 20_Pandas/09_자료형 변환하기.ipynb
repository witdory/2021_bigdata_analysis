# 9.1 데이터 타입 변환하기

```python
# 키 : float -> int
d['키'].dtype
```

```python
d['키'].fillna(-1,inplace=True)
d
```

```python
d['키'] = d['키'].astype(int)
```

```python
d['생년월일'].dtype
```

```python
# object -> datetime
d['생년월일']=pd.to_datetime(d['생년월일'])
d['생년월일']
```

```python
d['생년월일'].dt.year
```

```python
d['생년월일'].dt.month
```

```python
d['생년월일'].dt.day
```

```python
d['생일_년도']=d['생년월일'].dt.year
d['생일_월']=d['생년월일'].dt.month
d['생일_일']=d['생년월일'].dt.day
d
```
