## 10-2. DataFrame 병합하기(Merge)
* merge : 특정 키를 기준으로 병합하기 (예 : d1 과 d2 의 이름을 기준으로 병합하기)
* Database의 Join문과 유사
* pd.merge(left ,right ,on , how)
    * left, right에는 병합할 dataframe
    * on에는 병합의 기준이 되는 칼럼
    * ho2에는 'left', 'right', 'inner', 'outer' 4가지 병합 방식

### 1) left,right 방식

```python
# how='left'를 주면, left dataframe을 기준으로 병합이 됨(데이터 갯수 14개)
pd.merge(d1, d2, on='이름', how='left')
```

```python
# how='left'를 주면, left dataframe을 기준으로 병합이 됨(데이터 갯수 13개)
pd.merge(d1, d2, on='이름', how='right')
```

### 2) inner, outer 방식
* inner방식은 두 dataframe에 모두 키 값이 존재하는 경우에만 병합
* outer 방식은 하나의 dataframe에 키값이 존재하는 경우 모두 병합 -> 없는 값은 NaN으로 대입

```python
#inner : 데이터 갯수 12개
pd.merge(d1,d2,on='이름',how='inner')
```

```python
#outer : 데이터 갯수 15개
pd.merge(d1,d2,on='이름',how='outer')
```
