```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('seoul_temperature.csv',encoding='euc-kr')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('지점',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns=['날짜','평균','최저','최고']
df.head()
```

```python
df.날짜 = pd.to_datetime(df.날짜)
df.info()
```

```python
# 년도, 월, 일 컬럼 생성
df['년도'] = df.날짜.dt.year
df['월']=df.날짜.dt.month
df['일'] = df.날짜.dt.day
df.head()
```

# EDA (탐색적 데이터 분석)
* 평균 기온이 얼마나 상승했나?

```python
# 8월 1일 기준으로 매년 평균기온 분석
df_8=df[(df.월 == 8) & (df.일==1)]
df_8.head()
```

```python
# 1월 1일 기준으로 평균기온 분석
df_1 = df[(df.월==1) & (df.일==1)]
df_1.head()
```

```python
# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(10,5))
sns.lineplot(x='년도',y='평균',data=df_8)
sns.lineplot(x='년도',y='평균',data=df_1)
plt.show()
```

```python
df.head()
```

```python
## 가장 더웠던 날과 최고 기온은?
df.sort_values(by='최고',ascending=False)
```

```python
# 가장 추웠던 날과 최저 기온은?
df.sort_values(by='최저',ascending=True)
```

```python
# 일교차가 1년 중 가장 큰 달은?
df['일교차'] = df.최고 - df.최저
df.head()
```

```python
pd.DataFrame(df.groupby('월')['일교차'].mean())
```
