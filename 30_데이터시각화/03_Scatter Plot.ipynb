# 3. Scatter Plot
* 좌표상의 점들을 표시함으로써 두 변수간의 관계를 나타내줌

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50) # [0, 1)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
# Color 지정하기
plt.scatter(x,y,c='r')
plt.show()
```

```python
# 점마다 다른 color 지정하기
x = [1,2,3]
y = [1,2,3]
colors = ['r','g','b']
plt.scatter(x,y,c=colors)
plt.show()
```

```python
# ColorMap(색상지도)
# 데이터를 기바으로 색상 표현하기
plt.colormaps()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,c=x,cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# Marker Size 변경하기
size = np.random.rand(50) * 100 #0.1
plt.scatter(x,y,c=x,cmap='Blues',s=size)
plt.show()
```

```python
# 투명도(alpha, 0~1사이의 값)
plt.scatter(x,y,c=x,cmap='Blues',s=size,alpha=0.5)
plt.show()
```
