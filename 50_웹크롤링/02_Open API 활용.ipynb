### 보건복지부-코로나19 감염현황 조회 서비스 (XML)

```python
serviceKey = 'VvGRSt5XHZxim6y8kt5dCD2QRIseU3Ah%2F%2FpLpH0kSFNeHFHmivZB1gy9ikQosVsRi%2Bh%2Bx7Kx1B9hrXQwze43Uw%3D%3D'

url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?serviceKey={}&pageNo=1&numOfRows=10&startCreateDt=20210801&endCreateDt=20210830'.format(serviceKey)
url
```

```python
temp = xmltodict.parse(res.text)
temp
```

```python
result = json.loads(json.dumps(temp)) # OrderDitc ->dict(str) -> dict
result
```

```python
result['response']
```

```python
items = result['response']['body']['items']['item']

for item in items:
    print(item)
```

```python
# 일자(STATE_DT), 확진자수(DECIDE_CNT), 사망자수(DEATH_CNT), 치료중환자수(CARE_CNT)
state_dt = []
decide_cnt = []
death_cnt = []
care_cnt = []
for item in items:
    state_dt.append(item['stateDt'])
    decide_cnt.append(item['decideCnt'])
    death_cnt.append(item['deathCnt'])
    care_cnt.append(item['careCnt'])
state_dt
```

```python
df_data={
    '기준일':state_dt,
    '확진자수':decide_cnt,
    '사망자수':death_cnt,
    '치료중환자수':care_cnt
}
df_data
```

# 한국관광공사 - 관광정보서비스(json)

```python
serviceKey = 'VvGRSt5XHZxim6y8kt5dCD2QRIseU3Ah%2F%2FpLpH0kSFNeHFHmivZB1gy9ikQosVsRi%2Bh%2Bx7Kx1B9hrXQwze43Uw%3D%3D'
pageNo = 1
url = 'http://api.visitkorea.or.kr/openapi/service/rest/KorService/areaCode?ServiceKey={}&numOfRows=10&pageNo=1&MobileOS=ETC&MobileApp=AppTest&_type=json'.format(serviceKey)
url
```

```python
res = requests.get(url)
res
```

```python
res.text
```

```python
data = res.json()
data
```

```python
numOfRows = data['response']['body']['numOfRows']
totalCount= data['response']['body']['totalCount']
numOfRows,totalCount
```

```python
queryCount = totalCount // numOfRows
if totalCount % numOfRows>0:
    queryCount += 1
queryCount
```

```python
code = []
name = []
rnum = []

for i in range(queryCount): #0~1
    pageNo = i + 1 # 1~2
    url = 'http://api.visitkorea.or.kr/openapi/service/rest/KorService/areaCode?ServiceKey={}&numOfRows=10&pageNo=1&MobileOS=ETC&MobileApp=AppTest&_type=json'.format(serviceKey,pageNo)
    res = requests.get(url)
    data = res.json()
    
    items = data['response']['body']['items']['item']
    for item in items:
        code.append(item['code'])
        name.append(item['name'])
        rnum.append(item['rnum'])
name
```

```python
df_data = {
    'code':code,
    'name':name,
    'rnum':rnum
}

df = pd.DataFrame(data=df_data)
df
```
